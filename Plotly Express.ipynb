```python
import plotly.express as px
```

```python
import pandas as pd
```

```python
#Just like seaborn plotly has pre-installed datasets and the code in seaborn was sb.get_dataset_names()

# we can reference the dataset we want directly by adding its name as a method.
 #imporing the gapmider daatset
countries=px.data.gapminder()
```

```python
countries
```

```python
countries['year'].max()
```

```python
# creating a dataset of just the lastest year 2007
countries_07 = countries[countries['year']==countries['year'].max()]
```

```python
countries_07.head()
```

```python
# creating a dataset that only has values if the country is 'United Kingdom'
uk = countries[countries['country']=='United Kingdom']
```

```python
uk.head()
```

# Linechart

Pressing Shift+Tab inside the brackets brings up a dropdown menu with all the arguements



```python
px.line(data_frame=uk,x='year',y='pop')
```

If you hover around the plot you get a pop-up of the datavalue ast that point.

```python
# color is essentially hue in seaborn therefore pass in a column name not a color
px.line(data_frame=countries,x='year',y='pop', color='country')
```

If we click on a country it unchecks it. If we click again, it checks it back.

```python
px.line(data_frame=countries,x='year',y='pop', color='country')
```

# Areachart

Area charts are like line charts but with the space below shaded so that the line covers a certain area.

It, however, requires a line_group arguement to be satisfied. This set the metric the area below the line should use.

```python
px.area(data_frame=countries,x='year',y='pop',line_group='continent',color='country')
```

```python
px.area(data_frame=countries,x='year',y='pop',color='continent',line_group='country')
```

# Barchart

```python
px.bar(data_frame=countries_07,x='continent',y='pop')
```

```python
#coloring each continent by every country in it
px.bar(data_frame=countries_07,x='continent',y='pop',color='country')
```

# Scatterplot

```python
px.scatter(countries_07,'lifeExp','gdpPercap')
```

```python
# Let us use the facet_col arguement to facet by continents
px.scatter(countries_07,'lifeExp','gdpPercap',facet_col='continent')
```

```python
px.scatter(countries_07,'lifeExp','gdpPercap',facet_row='continent')
```

```python
# making use of the Shift+Tab to find arguements that might help us space out our plot
px.scatter(countries_07,'lifeExp','gdpPercap',facet_row='continent',height=800)
# the height and width arguements are specified in pixels
```

# Formatting plotly charts

## Adding Titles

https://plotly.com/python/figure-labels/#align-plot-title

Simple as just using the title arguement.

```python
px.scatter(countries_07,'lifeExp','gdpPercap',title='Life Expectancy vs GDP')
```

## Adding labels

Done by using the labels arguement and passing in a dictionary 

>*labels={'current_label_name': 'new_label_name'}.*

```python
px.scatter(countries_07,'lifeExp','gdpPercap',title='Life Expectancy vs GDP',labels={'lifeExp':'Life Expectancy','gdpPercap':'GDP per Capita'})
```

# 3D Charts

You can create 3-d charts in plotly. These charts take in an x,y, and z arguements.

## 3d scatterplot

*Syntax*

>px.scatter_3d(data_frame='dataset_name', x='',y='',z='')

<i> Loading in some data

```python
import plotly.express as px
import pandas as pd
```

```python
# tips dataset from Plotly Expres
tips = px.data.tips()
```

```python
# countries dataset from Plotly Expres
countries = px.data.gapminder()
```

```python
tips.head()
```

```python
countries.head()
```

```python
px.scatter_3d(data_frame=tips,x='total_bill',y='tip',z='size')
```

```python
# we can filter it by day and lets add some height
px.scatter_3d(data_frame=tips,x='total_bill',y='tip',z='size',color='day',height=600)
```

## 3-d linechart

```python
px.line_3d(data_frame=countries,x='gdpPercap',y='pop',z='year',width=900,height=700)
```

```python
px.line_3d(data_frame=countries,x='gdpPercap',y='pop',z='year',color='country',height=700)
```

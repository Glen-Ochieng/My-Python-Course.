# HTML
read_html: https://pandas.pydata.org/docs/reference/api/pandas.read_html.html

*Syntax* 



```python
# Reading a url of a page that contains multiple tables
url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_films'
```

The third table in this page : High grossing films by year of release is the table we want 



```python
# For offline jupyter-desktop, install this first
#!pip install lxml
```

```python
# Saving into an object called movie
movie=pd.read_html(url)

# urls are not file path, therefore, dont worry about changing the delimiters
```

```python
type(movie)
```

<b>Each element in this list should be a table, which we can index and it will come out as a dataframe.

```python
movie[0]
```

```python
type(movie[0])
```

Okay , moving on to the second table

```python
movie[1]
```

```python
movie[2]
```

```python
movie[2]
```

And finally, this table is the one we were looking for.

We can, hoewever, save time by inserting  a match arguement in the read_html code that has the table name we want.

```python
movie=pd.read_html(url,match='High-grossing films by year of release')
```

```python
# Now the first index becomes the table we want
movie[0]
```

```python
# We can save this output as a dataframe
movie=movie[0]
```

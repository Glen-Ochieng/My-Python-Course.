*Syntax*

>array_name[rowstart:rowstop:rowstep, columnstart:columnstop:columnstep]

*Remember to start indexing from 0.*

```python
array_2d
```

```python
# Let's get the last 2 row , from the third column to last in both.

array_2d[1:,2:]
```

```python
# Let's extract  the last element 50.
array_2d[2,6]
```

<b>Just how inserting -1 in the step parameter reverses the order of a list, doing it for a multi-dimensional array in stead reverses the ordering of rows as opposed to the elements of a row. To reverse the elements inside a row, means reversing the columns ordering which can be done by inserting -1 as the column-step parameter. 

<b>However, the syntax for reverse ordering for rows and columns differs.

<u>rows

> array_name[::-1,]

<U>columns

>array_name[:,::-1] NOT [,::-1]


<u> rows and columns

>array_name[::-1,::-1]

```python
array_2d
```

```python
array_2d[::-1,]
```

```python
array_2d[:,::-1] # The colons for the rows still must be included.
```

```python
array_2d[::-1,::-1]
```

<u> 7. np.zeros()

This is a function creates a new array filled with zeros. It's useful for initializing arrays with a specific shape and data type but filled entirely with zeros.

*Syntax*

>numpy.zeros(shape, dtype=float, order='C'/'Fortran')

Where:

* shape is the shape of the array, represented as a tuple of integers specifying the size of each dimension.
* dtype (optional) is the data type of the array elements. If not provided, the default data type is float.
* order (optional) specifies whether the array should be stored in row-major (C-style) or column-major (Fortran-style) order. The default is 'C' (row-major).

```python
zeros_array = np.zeros((2, 3))

print(zeros_array) # inserting print removes the word array that comes before the output
```

Where:

* shape is the shape of the array, represented as a tuple of integers specifying the size of each dimension.
* dtype (optional) is the data type of the array elements. If not provided, the default data type is float.
* order (optional) specifies whether the array should be stored in row-major (C-style) or column-major (Fortran-style) order. The default is 'C' (row-major).

#### <b>Array math functions

<u> 1.np.sum()

The Sum function can sum elements in an array or a given axis.



```python
array_3
```

```python
# If no axis is specified then it will sum all elements of array_3
np.sum(array_3)
```

```python
# Sum all elements of array_3 in the 0 axis i.e going downwards(this means each column is summed up)
np.sum(array_3,axis=0)
```

```python
# Sum all elements of array_3 in the 1 axis ie going sideways (this means each row is summed up)
np.sum(array_3, axis=1)
```

<u>2. np.cumsum()

The cumsum function takes the cumilative sum of elements along the given axis. If you provide no axis then its computed over a flattened array.

```python
array_3
```

```python
# Cummulative sum in the 1 axis{going sideways}
np.cumsum(array_3, axis=1)
```

```python
# Cummulative sum in the 0 axis(going downwards)
np.cumsum(array_3, axis=0)
```

```python
# When no axis is provided then the array is "flattened"
np.cumsum(array_3, axis=None)
```
